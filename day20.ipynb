```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, output_image_path):
    """
    Generates a WordCloud from the given text and saves it as an image.
    
    Args:
        text (str): The input text for the WordCloud.
        output_image_path (str): The file path to save the WordCloud image.
    """
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    wordcloud.to_file(output_image_path)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
text = 'data science machine learning artificial intelligence'
output_image_path = 'wordcloud.png'
generate_wordcloud(text, output_image_path)
print(f"WordCloud saved as {output_image_path}")
```
